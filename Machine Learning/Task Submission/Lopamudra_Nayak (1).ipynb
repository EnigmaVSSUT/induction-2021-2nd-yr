```python
 import numpy as np
 import matplotlib.pyplot as plt
 

```

Here we have input the library.


```python
from sklearn.datasets import load_iris
```

Here we input the data through Sklearn library.


```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
X = dataset.data
```

Data contains sepal length, sepal width, petal length and petal width.


```python
Y = dataset.target
```

Target contains their category.


```python
Y
```

```python
X
```

```python
plt.plot(X[:, 0][Y == 0]* X[:, 1][Y == 0], X[:, 1][Y == 0]* X[:, 2][Y == 0], 'r.', label='Setosa')
plt.plot(X[:, 0][Y == 1]* X[:, 1][Y == 1], X[:, 1][Y == 1]* X[:, 2][Y == 1], 'g.', label='Versicolor')
plt.plot(X[:, 0][Y == 2]* X[:, 1][Y == 2], X[:, 1][Y == 2]* X[:, 2][Y == 2], 'b.', label='Virginica')
plt.legend()
plt.show()
```

Here we have visualized the data.


```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

Here we have standardize the data.


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

We have divided our whole data into train and test cases using split.


```python
from sklearn.linear_model import LogisticRegression
```

LogisticRegression machine learning model is used to classify the three categories of flower.


```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train, Y_train)
```

```python
log_reg.score(X_test, Y_test)
```

Score function is used to check the accuracy of the model.
